# Data for spam emails

```python
df=pd.read_csv("/content/ML/Naive bias/spam.csv")
df.head()
```

```python
df.groupby("Category"). describe ()
```

```python
df["Spam"]=df["Category"].apply(lambda X: 1 if X=="spam" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train=train_test_split(df.Message,df.Spam,test_size=0.2)
```

#Now using count vectoring techniques

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_traincount=v.fit_transform(X_train.values)
X_traincount.toarray()[:3]
```

```python
#Now applying multinomial Naive bayes
#To the data
```

```python
from sklearn.naive_bayes import  MultinomialNB
model= MultinomialNB()
model.fit(X_traincount,Y_train)
```

Testing the model

```python
emails=["Ok","Upto 20% off on all items"]
emails_check = v.transform(emails)
predictions = model.predict(emails_check)
predictions
```

```python
X_testcount=v.transform(X_test)
model.score(X_testcount,Y_test)
```

#insted of calling contvector and Nb we will use pipline function

```python
from sklearn.pipeline import Pipeline

Clf=Pipeline ([
("Vectorizer",CountVectorizer ()),
("Nb",MultinomialNB ())])
```

```python
Clf.fit(X_train, Y_train)
Clf.score(X_test,Y_test)
```

```python
Clf.predict(emails)
```
