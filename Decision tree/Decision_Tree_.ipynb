```python
df=pd.read_csv("/content/ML/Decision tree/salaries.csv")
df.head()
```

```python
inputs=df.drop("salary_more_then_100k",axis=1)
```

```python
target=df["salary_more_then_100k"]
```

```python
inputs.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
# Creating a column or columns

inputs["company_n"]=le_company.fit_transform(inputs["company"])
inputs["job_n"]=le_job.fit_transform(inputs["job"])
inputs["degree_n"]=le_degree.fit_transform(inputs["degree"])
```

```python
inputs.head()
```

```python
input_n=inputs.drop(inputs[["company","job","degree"]],axis=1)
```

```python
input_n.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(input_n,target,test_size=0.2)
```

```python
len(X_train)
len(X_test)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
model.predict([[2,2,1]])
```

```python
model.score(input_n,target)
```
